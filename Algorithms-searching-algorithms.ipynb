## Breadth first search
Search through each vertex in a graph  
such that the algorith will  
search through every neighbors  
before move on to the next vertex.

### Pseudocode
**Input**:  
a graph `g` and a starting vertex `v`  
**Output**:  
an order of g.nodes given by the breadth first search,  
starting at `v`

```Python
searched = [v]
waiting = [v]
while len(waiting) != 0:
    anchor = waiting[0]
    remove anchor from waiting
    
    new = [neighbors of anchor that is not yet searched]
    searched += new
    waiting += new
```

##### Sample implementation of BFS

````python
def BFS(g, v):
    """
    Input: 
        g: a graph (in NetworkX)
        v: a vertex
    Output: 
        an order of g.nodes given by the breadth first search,  
        starting at v.
    """
    searched = [v]
    waiting = [v]
    while len(waiting) != 0:
        anchor = waiting.pop(0)
        new = [u for u in g.neighbors(anchor) if u not in searched]
        searched += new
        waiting += new
    return searched
````
